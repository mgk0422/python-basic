## 1. countplot
- 데이터 수를 보고싶을때
- y축은 항상 count 이기때문에 y축을 따로 지정할수 없다.
- hue
- 컬럼  value의 종류란? ex) 'Survived'컬럼은 0과 1로 value가 두 종류

```python
sns.countplot(data=data,x='Sex')
```

```python
sns.countplot(data=data,x='Sex',hue='Survived')
```

## 2. distplot

## 3.barplot
- x별 y의 평균치를 보여준다.

```python
data.head(3)
```

```python
sex_age_mean=data.groupby('Sex')[['Age']].mean()
sex_age_mean
```

```python
sex_age_mean.plot.bar()
plt.grid()
plt.show()
```

```python
sex_survived_mean=data.groupby('Sex')[['Survived']].mean()
sex_survived_mean
```

```python
merge_sex_age_survived=pd.merge(sex_age_mean,sex_survived_mean,on='Sex')
merge_sex_age_survived
```

```python
merge_sex_age_survived.plot.bar()
```

## 4. boxplot
- 박스의 정중앙이 (검정선과 별개로) 평균값(mean)
- 검정선은 중앙값 (median) 
- 평균은 말그대로 n분의 1로 평균을 낸 것이고, 중앙값은 숫자를 1~100까지 나열했을 때 중앙에 있는 값.

```python
class_age_df=data[['Pclass','Age']]
class_age_df
```

```python
class_age_df[['Pclass','Age']].boxplot(by='Pclass')
```

```python
# 나이(Age)가 50살이상 60살 이하인 성별(Sex) 평균요금(Fare)
age5060=data.query('Age >=50 and Age<=60') 
age5060
```

```python
sex_fare_mean=age5060.groupby('Sex')[['Fare']].mean()
sex_fare_mean
```

```python
sex_fare_mean.plot.bar()
```
