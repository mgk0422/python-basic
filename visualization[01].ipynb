```python
# subplot(2, 1, 1)
# subplot(2, 1, 2)

X1 = np.linspace(0.0, 5.0)
Y1 = np.cos(2 * np.pi * X1) * np.exp(-X1)
print(X1)
print(Y1)

X2 = np.linspace(0.0, 2.0)
Y2 = np.cos(2 * np.pi * X1)
print(X2)
print(Y2)
```

```python
axes01=plt.subplot(2,1,1)
plt.plot(X1,Y1)
plt.title('subplot 01')
axes02=plt.subplot(2,1,2)
plt.plot(X2,Y2)
plt.title('subplot 02')

plt.tight_layout() # 자동으로 간격을 맞춰주는 역할
plt.show()
```

```python
# bar plot
# x 데이터는 카테고리 값인 경우가 대부분이다.

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

Y=[2,3,1]
X=np.arange(len(Y))

xlabel=['1등실','2등실','3등실']
plt.title('bar plot')
# plt.barh(X,Y) # 수평형태로
plt.bar(X,Y)
plt.xticks(X,xlabel)
plt.yticks(sorted(Y))
plt.xlabel('선실 등급')
plt.ylabel('생존자 수')
plt.show()
```

```python
# bar plot
# x 데이터는 카테고리 값인 경우가 대부분이다.

import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline

Y=[2,3,1]
X=np.arange(len(Y))

ylabel=['1등실','2등실','3등실']
plt.title('bar plot')
plt.barh(X,Y) # 수평형태로
plt.xticks(sorted(X))
plt.yticks(Y,ylabel)
plt.xlabel('생존자수')
plt.ylabel('선실등급')
plt.show()
```

- 데이터 전처리 : 연료가격 변수 추가
- c :가스 , d :디젤, e:에탄올 , p : 고급휘발유, F:보통휘발유

```python
fl_df=pd.DataFrame({
    'fl':['c','d','e','p','r'],
    'price_fl':[2.35,2.38,2.11,2.76,2.22]
})
fl_df
```

```python
# 우리가 만든 연료 가격 프레임을 원본 프레임에 병합 
data_fl_maerge_df=pd.merge(data_df,fl_df,on='fl',how='inner')
data_fl_maerge_df.head(5)
```

```python
# 머지한 데이터프레임에서  model,fl,price_fl 만 추출
data_fl_maerge_df[['model','fl','price_fl']]
```

```python
data_fl_maerge_df.isna().sum()
```

```python
# 구동방식(div)별 고소고도로 연비(hwy)평균
# 임의적으로 결측값 처리를 위해 더미 값을 넣어보도록 하자
data_fl_maerge_df.loc[65,'hwy']=np.nan
data_fl_maerge_df.loc[120,'hwy']=np.nan
data_fl_maerge_df.loc[154,'hwy']=np.nan
data_fl_maerge_df.loc[178,'hwy']=np.nan
data_fl_maerge_df.loc[219,'hwy']=np.nan
data_fl_maerge_df.loc[230,'hwy']=np.nan
```

```python
data_fl_maerge_df['drv'].value_counts()
```

```python
# hwy 변수의 결측값을 제외하고, 어떤 구동방식의 고속도로평균 연비가 높은지 알아보자.
data_fl_merge_isna_df=data_fl_maerge_df.filter(['drv','hwy']).dropna()
drv_group_df=data_fl_merge_isna_df.groupby('drv')[['hwy']].mean()
display(drv_group_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data_fl_maerge_df.head(5)
```

```python
# 구동방식별 연비평균을 비교하기 위한 막대 그래프를 시각화 해보자
drv_group_df.plot.bar(rot=0)
plt.title('구동방식별 평균연비') 
plt.xlabel('구동방식') 
plt.ylabel('평균연비')
plt.show()
```

```python
# 위에서 했던 방식과 동일하게 구동방식별 고속도로, 도시연비의 평균을 구해보고
# 이를 데이터 프레임으로 만들어서 막대바로 시각화 해보기
data_fl_maerge_df.head(5)
cty_group_df=data_fl_maerge_df.groupby('drv')[['cty']].mean()
cty_group_df
merge_df=pd.merge(drv_group_df,cty_group_df,on='drv',how='inner')
merge_df
merge_df.plot.bar(rot=0)
plt.grid()
```

```python
# 자동차(class)중에서 어떤 자동차가 가장 많은지 알아보려고 한다.
# 종류별 빈도를 막대바로 시각화해보기
car_count=data_fl_maerge_df['class'].value_counts()
car_count.plot.bar(rot=0)
plt.title('자동차 종류 빈도') 
plt.xlabel('자동차 종류') 
plt.ylabel('count')
plt.grid()
```

```python
#  어떤 회사에서 생산한  suv차종의 도시 연비가 높은지 알아보려고 한다.
# suv 차종을 대상으로 평균 도시연비가 가장 높은 회사 다섯곳을 막대 바로 시각화 해보자.
suv_df=data_fl_maerge_df[data_fl_maerge_df['class']=='suv']
cty_df=suv_df.groupby('manufacturer')[['cty']].mean()
cty_df.sort_values(by='cty',ascending=False,inplace=True)
cty_df=cty_df.iloc[0:5,:]
cty_df.plot.bar(rot=0)
plt.title(' 어떤회사에서 생산한 suv차종의 도시 연비가 높을까 ') 
plt.xlabel('suv차종') 
plt.ylabel('평균 도시연비')
plt.grid()
```
