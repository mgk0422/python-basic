```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as pltaa
```

```python
def serieinfo(s):
    print('value :',s.values)
    print('value type :',type(s.values))
    print('index :',s.index)
    print('index type:',type(s.index))
    print('index+value:',s)
```

```python
price_series=pd.Series([14000,3000,3500,2000],
                      index=['a','b','c','d'])
serieinfo(price_series)
```

```python
price_series[0]=4000
serieinfo(price_series)
```

```python
# 값 추가
price_series['e']=1000
serieinfo(price_series)
```

```python
# 삭제
del price_series['e']
serieinfo(price_series)
```

```python
set=pd.Series(list({10,20,30,40,50})) # list를 통해 순서가 잇는 형식으로 만들어야함
serieinfo(set)
```

```python
pd.isnull(set) # null값이 있다없다 판단
```

```python
# 널 값을 넣기위해서 numpy에서  제공하는 np.NaN 사용해야한다
set[0]=np.NaN
serieinfo(set)  #float 형식으로 바뀜..!
```

```python
ser01=pd.Series([100,200,300,350],index=['a','o','k','m'])
ser02=pd.Series([150,100,250,450],index=['o','h','k','a'])
```

```python
def dfinfo(df):
    print('df shape: {}'.format(df.shape))
    print('df size : {}'.format(df.size))
    print('df ndim : {}'.format(df.ndim))
    print('df index : {}'.format(df.index))
    print('df index type :{}'.format(type(df.index)))
    print('df columns :{}'.format(df.columns))
    print('df cloumns :{}'.format(type(df.columns)))
    print('df values:{}'.format(df.values))
    print('df values type:{}'.format(type(df.values)))
```

```python
score_data={
    'kor':[80,90,70,30],
    'eng':[90,70,60,40],
    'math':[90,60,90,40]
}
columns=['kor','eng','math']
index=['김지은','황인범','김정수','최호진']
exec_df=pd.DataFrame(score_data,index=index,columns=columns)
exec_df
```

- 위 데이터를 보고 모든 학생의 수학 점수를 시리즈로 출력하라
- 모든 학생의 국어와 영어 점수를 데이터 프레임으로 만들어라
- 모든 학생의 각 과목 평균 점수를 새로운 열로 추가하라
- 최호진 학생의 영어 점수를 90점으로 수정하고 평균 점수도 다시 계산하라
- 김지은 학생의 점수를 데이터 프레임으로 만들어라
- 김정수 학생의 점수를 시리즈로 출력하라
- 황인범 학생의 국어 점수와 수학 점수를 100점으로 수정하고 평균 점수도 다시 계산하라

```python
# 위 데이터를 보고 모든 학생의 수학 점수를 시리즈로 출력하라
exec_df['math']
```

```python
# 모든 학생의 국어와 영어 점수를 데이터 프레임으로 만들어라
exec_df[['kor','math']]
```

```python
# 모든 학생의 각 과목 평균 점수를 새로운 열로 추가하라
exec_df['Avg'] = ((exec_df['kor'] + exec_df['eng'] + exec_df['math']) / 3).round(2)
exec_df
```
