```python
# 파일 읽기

seoul_uni_df=pd.read_excel('./data/서울지역 대학교 위치.xlsx')
seoul_uni_df.head()
```

```python
seoul_map=g.Map(location=[37.55,126.98],
                tiles='Stamen Terrain',
                zoom_start=12)
# 대학교 위치정보를 marker를 표시
for idx in range(seoul_uni_df.shape[0]):
    #print(x)
    marker01=g.Marker(seoul_uni_df.iloc[idx,[0,1]],
                      popup=seoul_uni_df.index[idx],
                      icon=g.Icon(color='blue'))
    marker01.add_to(seoul_map)
seoul_map
```

```python
seoul_map=g.Map(location=[37.55,126.98],
                tiles='Stamen Terrain',
                zoom_start=12)

for name,lat,lng in zip(seoul_uni_df.index,seoul_uni_df.위도,seoul_uni_df.경도):
    g.CircleMarker([lat,lng],popup=name,icon=g.Icon(color='red'),
    radius=10,fill_color='coral',fill_opacity=0.3).add_to(seoul_map)
    
seoul_map
```

```python
elem_school_df=pd.read_excel('./data/elem_school_folium.xlsx')
elem_school_df.head(10)
```

```python
# 학교명,학교급구분,소재지도로명,위도,경도
elem_school_df=elem_school_df[['학교명','학교급구분','소재지도로명주소','위도','경도']]
```

```python
elem_school_df.dropna(axis=0,inplace=True)
elem_school_df.isnull().sum()
```

```python
# 문자열 다루는 방법. str[:]
# 소재지도로명주소가 서울을 포함하고 학교급 구분이 초등학교만 추출
elem_school_df=elem_school_df[(elem_school_df['소재지도로명주소'].str.contains('서울')) & (elem_school_df['학교급구분']=='초등학교')]
elem_school_df
```

```python
elem_school_df.shape
```

```python
seoul_map=g.Map(location=[37.55,126.98],
                tiles='Stamen Terrain',
                zoom_start=12)

for name,lat,lon in zip(elem_school_df.학교명, elem_school_df.위도, elem_school_df.경도):
    g.Marker([lat, lon], 
             popup=name, 
             icon=g.Icon(color='red')).add_to(seoul_map)
seoul_map
```
