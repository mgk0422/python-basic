- pie chart : 카테고리별 값의 상대적 비교를 해야할 경우
- 차트 원형을 유지 : plt.axis('equal')

```python
labels=['뽕사부','순대국밥','개구리비빔밥','생고기','지히철버거']
sizes=[15,15,20,25,25]
colors=['green','gold','blue','red','yellow']
plt.title('Pie Chart')
plt.pie(sizes,labels=labels,colors=colors,shadow=True,autopct='%1.1f%%')
plt.axis('equal') # 모양유지
plt.show()
```

- histogram : 구간에 대한 데이터 집계를 시각화 할 때

```python
data=np.random.randn(1000)
plt.title('histogram')
plt.hist(data,bins=20) #bins=20 막대사이즈! 숫자가 커질수록 밀집하게 나타남
plt.show()
```

- scatter plot : 2차원 데이터
- 데이터 집합의 상관관계를 살펴보기위한

```python
np.random.seed(100)
X=np.random.normal(0,1,100)
Y=np.random.normal(0,1,100)
plt.title('scatter plot')
plt.scatter(X,Y)
plt.show()
```

```python
np.random.seed(100)
X=np.random.rand(30)
Y1=np.random.rand(30)
Y2=np.random.rand(30)
Y3=np.pi*(15*np.random.rand(30)**2)
plt.title('bubble chart')
plt.scatter(X,Y1,c=Y2,s=Y3)
plt.show()
```

```python
import seaborn as sns
iris=sns.load_dataset('iris')
print(iris.head())
titanic=sns.load_dataset('titanic')
```

```python
iris.sepal_length[:20].plot(kind='bar',rot=0)
plt.show()
```

```python
# iris[:5].plot(kind='bar',rot=0)
iris[:5].plot.bar(rot=0)
plt.show()
```

```python
# 그룹분석으로 각 붓꽃종의 평균을 구하고 시각화한다.

iris_bar_df=iris.groupby('species')[['sepal_length' ,'sepal_width','petal_length','petal_width']].mean()
iris_bar_df.columns.name='feature'
iris_bar_df
```

```python
iris_bar_df.plot.bar(rot=0)
plt.show()
```

```python
iris_bar_df.T.plot.bar(rot=0)
plt.show()
```

- pie,hist,box,scatter

- autopct 사용법
- '%0.2f' — 소숫점 두번째자리 까지 보여준다. 4.08
- '%0.2f%%' — 소숫점 두번째자리까지 보여주고 맨 뒤에 ‘%’ 기호를 붙인다. 4.08%.
- (% 기호를 두 번 쓰는 이유는 문자열 포매팅 할 때 처음 % 기호가 이스케이프 문자처럼 작동하기 때문이다.)
- '%d%%' — 가장 가까운 정수로 반올림하고 맨 뒤에 % 기호를 붙인다.

```python
pclass_count=titanic.groupby('pclass')['count'].sum()
pclass_count
```

```python
plt.title('Pie Chart')
pclass_count.plot(kind='pie',
                  shadow=True,
                  autopct='%1.1f%%',
                  legend=True,
                  startangle = 45)
plt.axis('equal') # 모양유지
plt.show()
```
