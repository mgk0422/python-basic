```python
import matplotlib.pyplot as plt
plt.figure() # ffogrure은 그래프를 표현할 틀을 만든다
plt.show() # figure을 출력한다.
```

#### 선그래프

```python
plt.figure()
# (1,1),(2,2),(3,3),(4,4)의 데이터를 넣어서 그린다.
# plt.plot([x축 데이터],[y축 데이터])
plt.plot([1,2,3,4],[1,2,3,4])
plt.show()
```

```python
# 찍은점을 표현하고 싶다면? marker인자 사용하기

plt.figure()
plt.plot([2,4,6,8],[10,20,30,40],marker="o")
plt.show()
```

#### 스캐터 그래프

```python
plt.figure()
# plt.scatter([x축 데이터],[y축 데이터])

#(1,1),(2,2),(3,3)에 점을 찍은 그래프를 그린다
plt.scatter([1,2,3],[1,2,3])
plt.show()
```

#### 막대그래프

```python
# plt.bar([x축 데이터],[y축 데이터])

plt.figure()
plt.bar([1,2,3,4,5,6,7,8],[15,2,4,9,1,12,14,10])
plt.show()
```

#### 파이그래프

```python
plt.figure()
# 데이터 비율이 1:2:3인 파이 그래프를 그린다.
plt.pie([1,2,3])
plt.show()
```

### 여러개의 그래프 그리기

```python
# 한번에 여러개의 그래프를 그리려면 plt.subplots(행,열)를 이용하면 된다.
# plt.subplots(행,열)은 2개의 변수를 변환하는데
# 하나는 전체 액자 figure에 대한 변수이고,
# 다른 하나는 액자 내 여러개의 액자에 대한 리스트 axs이다.

fig,ax=plt.subplots(1,3)
plt.show()
```

```python
# 1행 3열로 액자를 그린다.
fig,ax = plt.subplots(1,3)

# 각 액자들에 그래프를 그린다.
ax[0].plot([1,2,3],[1,2,3])
ax[1].bar([1,2,3],[1,2,3])
ax[2].pie([1,2,3])
plt.show()
```

#### 그래프 크기 변경하기

```python
# 가로 10, 세로 5인 틀을 만든다

plt.figure(figsize=(10,5))
plt.plot([1,2,3,4],[10,20,30,40],marker="o")
plt.show()
```

```python
# 여러개의 그래프를 그릴 때로 figsize를 추가해주면된다. 

# 1행 3열로 액자를 그린다. 가로 20 세로 5
fig,ax = plt.subplots(1,3,figsize=(20,5))
 
# 각 액자들에 그래프를 그린다. # 컬러변경
ax[0].plot([1,2,3],[1,2,3],color="green")
ax[1].bar([1,2,3,4],[175,158,174,163],color='red')
ax[2].pie([1,2,3])
plt.show()
```

```python
# 제목, 축이름 붙이기

# 방법1. plt에 직접 붙이기

# 그래프 제목 : plt.title()
# 각 축의 이름 : plt.xlabel(),plt.ylabel()
# 한글 꺠지는 것 

# 폰트 설정 방법 1
import matplotlib.pyplot as plt

# matplotlib 폰트설정
plt.rc('font', family='NanumGothic') # For Windows
print(plt.rcParams['font.family'])

plt.figure()
plt.bar([1,2,3,4],[175,158,174,163])
plt.title("가족 키 그래프")
plt.xlabel("키")
plt.ylabel("가족")
plt.show()
```

```python
# 방법2. ax에 설정

# 시각화 코드들을 보다보면, plt.figure() 을 사용하지 않고, 
# 주로 plt.subplots()을 사용하는 경우가 많은데,
# 이 때 ax를 이용하여 설정할 수 있다.

fig, ax = plt.subplots(1, 1)
ax.plot([1,2,3,4,5], [1,3,5,7,9])
ax.set_title("제목")
ax.set_xlabel("x축 이름") 
ax.set_ylabel("y축 이름")
plt.show()
```

```python
# 각각 그래프에서 설정하기

fig, ax = plt.subplots(1, 2, figsize=(20, 5))
# 0번 그래프
ax[0].bar([1,2,3,4,5,6,7], [15,28,95,56,41,85,87], color='red') 
ax[0].set_title("A반 수학점수")
ax[0].set_xlabel("학생 번호")
ax[0].set_ylabel("수학 점수") 

# 1번 그래프
ax[1].bar([1,2,3,4,5,6,7], [42,45,100,96,75,65,43], color='green')
ax[1].set_title("B반 수학점수")
ax[1].set_xlabel("학생 번호")
ax[1].set_ylabel("수학 점수")

plt.show()
```

```python
# 축 단위 바꾸기
# plt.xticks([단위 리스트])를 사용하면 x축 단위로 수동으로 바꿀수 있다

plt.figure()
plt.plot([1,2,3],[1,2,3])
plt.xticks([1,2,3],['월','수','금']) # x축 단위 바꾸기
plt.yticks([1,2,3],['저','중','고']) # y축 단위 바꾸기
```

```python
# 축범위 바꾸기
# plt.xlim([보여줄 최소값, 보여줄 최대값])

plt.figure()
plt.plot([1,2,5],[10,20,25])
plt.xlim([0,10]) # x축 범위 바꾸기
plt.ylim([5,50]) # y축 범위 바꾸기
plt.show()
```

```python
# 그래프가 여러개인경우?

fig,ax=plt.subplots(1,2,figsize=(20,5))

# 0번 그래프
ax[0].plot([1,2,3],[1,2,3])
ax[0].set_xticks([1,2,3])

# 1번 그래프
ax[1].plot([1,2,3],[1,2,3])
ax[1].set_xlim([0,5])

plt.show()
```

```python
# 범례(legend)표시
# 범례를 표시하려면, plt.plot의 인자로 label을 설정해주고
# plt.legend()를 적어주면 된다

plt.figure()
plt.plot([1,2,3],[1,2,3],label="그래프")
plt.legend()
plt.show()
```
