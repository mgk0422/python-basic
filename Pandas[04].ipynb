#### 데이터프레임 인덱스 조작하는 방법
- set_index : 인덱스를 세팅해줌 : 기존 행 인덱스를 제거하고 데이터 열 중 하나를 인덱스로 설정

```python
# 5명의 학생의 국어, 영어, 수학 점수를 나타내는 데이터프레임을 만든다.

# 1. 
# 학생 이름을 나타내는 열을 포함시키지 않고 
# 데이터프레임 df_score1 을 생성한 후, 
#df_score1=score.set_index('국어')
#df_score1

# df_score1.index 속성에 학생 이름을 나타내는 열을 지정하여 인덱스를 지정한다. 
# reset_index 명령으로 이 인덱스 열을 명령으로 
# 일반 데이터열로 바꾸여 데이터프레임 df_score2을 만든다.

df_score1=pd.DataFrame({
    '국어':[100,50,21,45,100],
    '영어':[80,42,42,65,45],
    '수학':[100,85,45,25,95],
})
df_score1.index=['민경','우석','지은','세진','지원']
df_score1.index.name='이름'
df_score2=df_score1.reset_index()
display(df_score2)

# 2.
# 학생 이름을 나타내는 열이 일반 데이터 열을 포함하는 데이터프레임 df_score2에 
# set_index 명령을 적용하여 다시 학생 이름을 나타내는 열을 인덱스로 변경한다.

df_score3=df_score2.set_index('이름')
df_score3
```

#### DataFrame merge

```python
data1 = {
    '학번' : [1,2,3,4],
    '이름' : ['섭섭해','김한준','김선림','최호진'],
    '학년' : [2,4,1,3]
}

data2 = {
    '학번' : [4,5,2,1],
    '학과' : ['CS','MATH','MATH','CS'],
    '학점' : [2.4, 4.5, 3.4, 4.2]
}

# 공통의 열은 학번!!!!!!!!!!! 
# merge는 두개의 프레임을 합치는건데  row의 인덱스로 합칠수 있고, 열로 합칠수 있다.!

# 공통의 컬럼이 있을경우 inner 쓸수 있고
# 공통의 컬럼이 없을경우 merge 해야한다!!1!
```

```python
stu_df=pd.DataFrame(data1)
major_df=pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
pd.merge(stu_df,major_df) #학번을 기준으로 매칭된다! 매칭되지않는것은 누락된다!
```

```python
pd.merge(stu_df,major_df,on='학번',how='inner') # on='학번',how='inner' 이거 없어도 동일 디폴트값을 의미 
# on은 공통컬럼
```

```python
pd.merge(stu_df,major_df,on='학번',how='left')
```

```python
pd.merge(stu_df,major_df,on='학번',how='right')
```

```python
pd.merge(stu_df,major_df,on='학번',how='outer') # 전체 데이터 출력
```

- 컬럼의 인덱스가 다른 경우 merge

```python
data1 = {
    '학번' : [1,2,3,4],
    '이름' : ['섭섭해','김한준','김선림','최호진'],
    '학년' : [2,4,1,3]
}

data2 = {
    '과목코드' : [4,5,2,1],
    '학과' : ['CS','MATH','MATH','CS'],
    '학점' : [2.4, 4.5, 3.4, 4.2]
}
```

```python
stu_df=pd.DataFrame(data1)
major_df=pd.DataFrame(data2)
display(stu_df)
display(major_df)
```

```python
pd.merge(stu_df, major_df,left_on='학번',right_on='과목코드',how='inner')
```

- left_index, right_index

```python
pop_df01=pd.DataFrame({
    'city':['seoul','seoul','seoul','busan','busan'],
    'year':[2010,2005,2020,2018,2015],
    'pop':[1234567,2345678,3456789,4567890,5678901]
                    })
pop_df01
```

```python
pop_df02    = pd.DataFrame( np.arange(12).reshape((6,2)) , 
    index   = [['busan' , 'busan', 'seoul' , 'seoul' , 'seoul' , 'seoul'] , 
               [2010, 2005, 2020, 2018, 2015, 2010]] , 
    columns =  ['col01','col02']
)
pop_df02
```

```python
pd.merge(pop_df01,pop_df02,left_on=['city','year'], right_index=True)
```

- 다중 인덱스가 아닌 단일 인덱스를 이용하는 방법

```python
data1 = { "이름" : ["이지안","박동훈","이순신","강감찬"],
          "학년" : [2,4,1,3]}

data2 = { "학과" : ["CS","MATH","MATH","CS"],
          "학점" : [3.4,2.9,4.5,1.2]}
```

```python
df01=pd.DataFrame(data1,index=[1,2,3,4])
df02=pd.DataFrame(data2,index=[1,2,4,5])
```

```python
display(df01)
display(df02)
```

```python
merge_df=pd.merge(df01,df02,right_index=True,left_index=True) # right_index=True,left_index=True 행인덱스를 사용한다! 
```
